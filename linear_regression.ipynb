```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean(1).csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6)) 
plt.scatter(X,y, alpha = 0.3) #o alpha é para transparência dos pontos
plt.title('Film cost vs Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000 #limite mínimo e máximo dos valores de y
plt.xlim(0, 450000000) #limite mínimo e máximo dos valores de x
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
#Slope Coefficient
regression.coef_ #theta_1, ou seja, hipótese 1
```

```python
#Intercept, o theta 0
regression.intercept_ 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha = 0.3)
#Adicionar linha no mesmo gráfico usando o regression.predict()
plt.plot(X, regression.predict(X), color='red', linewidth=4)

plt.title('Film cost vs Global Revenue')
plt.xlabel('Production budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
 regression.score(X,y) #esse valor representa a porcentagem dos dados que a previsão da regressão linear acerta
#neste caso, a regressão linear prevê bem para 0.54, ou seja, para 54% dos casos
```
